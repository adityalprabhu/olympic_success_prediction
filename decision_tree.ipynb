```python
import pandas as pd
from sklearn import tree
import graphviz
import pydotplus
```

```python
data = pd.read_table('data/olympics.csv', sep=',', header=0, engine='python')
```

```python
data
```

```python
data = data.replace({'Sex' : { 'F' : 0, 'M' : 1}})
data = data.replace({'Season' : { 'Summer' : 0, 'Winter' : 1}})
temp = data.pop('Medal')
data['Medal']=temp
sports = data['Sport'].unique()
sport_label = dict(zip(sports, range(len(sports))))
data['Sport'] = [sport_label[s] for s in data['Sport']]
```

```python
feature_cols =['Sex', 'Age','Height','Weight', 'Season', 'Sport','Average Temperature', 'GDP Per Capita']

target = 'Medal'

x = data[feature_cols]

y = data[target]

clf = tree.DecisionTreeClassifier()
clf = clf.fit(x,y)
```

```python
dot_data = tree.export_graphviz(clf, out_file=None, 
                      feature_names=feature_cols,  
                      class_names=target,  
                      filled=True, rounded=True,
                      special_characters=True)  
graph = graphviz.Source(dot_data)  
```
