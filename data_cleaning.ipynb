```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
climate_data = pd.read_table('data/scraped_climate_info.csv', sep=',', header=0, engine='python')
```

```python
gdp_data = pd.read_table('data/gdpdata.csv', sep=',', header=0, engine='python')
```

```python
olympics_data = pd.read_table('data/athlete_events.csv', sep=',', header=0, engine='python')
```

```python
climate_data.head()
```

```python
gdp_data.head()
```

```python
olympics_data.head()
```

```python
gdp_data.rename(columns={'Name':'Country'}, inplace=True)
```

```python
common_countries = pd.merge(climate_data, gdp_data, how='inner', on=['Country'])
climate_gdp = common_countries[['Country','Average Temperature', 'GDP Per Capita']]
```

```python
climate_data.shape
```

```python
gdp_data.shape
```

```python
common_countries.head()
```

```python
climate_gdp.head()
```

```python
climate_gdp['Average Temperature'] = climate_gdp['Average Temperature'].str.replace('−', '-').astype(float)
```

```python
olympics_data = olympics_data.replace({'Medal' : { 'Gold' : 1, 'Silver' : 2, 'Bronze' : 3}})
olympics_data['Medal'] = olympics_data['Medal'].fillna(0)
```

```python
olympics_data.shape
```

```python
olympics_data.rename(columns={'Team':'Country'}, inplace=True)
```

```python
olympics_data = olympics_data.dropna()
```

```python
olympics_data = olympics_data[olympics_data['Year'].between(1980, 2018, inclusive=True)].sort_values(by=['Year'])
```

```python
olympics = pd.merge(olympics_data, climate_gdp, how='inner', on=['Country'])
olympics.shape
```

```python
olympics_csv = olympics.to_csv (r'data/olympics.csv', index = None, header=True) 
```
